# Implementing switching methods

## Switching

```python
def switch(G: Graph, Q: matrix, C: list):
    A = G.adjacency_matrix(vertices=C+[x for x in G if not x in C])
    QI = block_diagonal_matrix([Q,identity_matrix(len(G)-Q.nrows())])
    return Graph(QI.T*A*QI)
```

```python
def blocks(Q): # subsets of {0,...,n-1} whose characteristic vectors are the 01-vectors x such that Q^T*x is a 01-vector again
    L = []
    n = Q.nrows()
    for x in GF(2)^n:
        x = vector(QQ(i) for i in x)
        if all(i==0 or i==1 for i in Q.T*x):
            L.append({i for i in range(n) if x[i] == 1})
    return L
```

```python
def allSwitchingSets(G: Graph, switching_subgraph: Graph, blocks: list):
    assert set(switching_subgraph.vertices()) == set(range(switching_subgraph.order()))
    assert (all(p in switching_subgraph.vertices() for p in B) for B in blocks)
    
    H = G.disjoint_union(switching_subgraph)
    for b, B in enumerate(blocks):
        H.add_vertex(b)
        H.add_edges(((1, p), b) for p in B)
    vertices = {(0,v) for v in G}
    points = {(1,p) for p in switching_subgraph}
    blockset = list(range(len(blocks)))
    for switchingSet, ordering in chooseRecursively(H, [], vertices, [], points, blockset, {b:[] for b in blockset}, {v:[] for v in vertices}, {p:vertices for p in points}):
        yield [switchingSet[ordering.index((1, i))][1] for i in range(switching_subgraph.order())] # ordered switching set
```

```python
def chooseRecursively(H, chosenVertices, availableVertices, chosenPoints, availablePoints, blockset, blockpatterns, vertexpatterns, candidates):
    
    aut = H.automorphism_group(partition=[chosenVertices, availableVertices, chosenPoints, availablePoints, blockset])
    
    chosenMatches = [(chosenVertices[i],chosenPoints[i]) for i in range(len(chosenVertices))]
    availableMatches = [(v,p) for v in availableVertices for p in availablePoints]
    chosenMatches_orbits = DisjointSet(chosenMatches)
    availableMatches_orbits = DisjointSet(availableMatches)
    for f in aut.gens():
        for (v,p) in chosenMatches:
            chosenMatches_orbits.union((v,p), (f(v),f(p)))
        for (v,p) in availableMatches:
            availableMatches_orbits.union((v,p), (f(v),f(p)))
    
    if chosenVertices:
        # first heuristic
        degrees = [len(set(chosenPoints).intersection(H.neighbors(p))) for p in chosenPoints]
        unique_degrees = [d for d in degrees if degrees.count(d) == 1]
        if unique_degrees and not degrees[-1] == min(unique_degrees):
            # not canonical
            return
        # canonical labeling of entire graph
        _,Hcanonical = H.canonical_label(partition=[chosenVertices, availableVertices, chosenPoints, availablePoints, blockset], certificate=True)
        canonical_vertex = min(chosenVertices, key=lambda v: Hcanonical[v])
        canonical_match = (canonical_vertex,chosenPoints[chosenVertices.index(canonical_vertex)])
        if chosenMatches_orbits.find((chosenVertices[-1],chosenPoints[-1])) != chosenMatches_orbits.find(canonical_match):
            # not canonical
            return
    
    if len(availablePoints) == 0:
        yield chosenVertices, chosenPoints
    
    for (v,p) in availableMatches:
        if availableMatches_orbits.find((v,p)) != (v,p) or any(H.has_edge(v, chosenVertices[i]) != H.has_edge(p, chosenPoints[i]) for i in range(len(chosenVertices))):
            continue
        chosenVertices.append(v)
        availableVertices.remove(v)
        chosenPoints.append(p)
        availablePoints.remove(p)
        H.add_edge(v,p)
        # update blockpatterns
        for b in blockset:
            blockpatterns[b].append(H.has_edge(p, b))
        # update vertexpatterns
        countmustadd = 0
        for w in availableVertices:
            vertexpatterns[w].append(H.has_edge(v, w))
            if vertexpatterns[w] not in blockpatterns.values():
                countmustadd += 1
        if countmustadd <= len(availablePoints):
            newcandidates = {}
            for q in availablePoints:
                if q in H.neighbors(p):
                    newcandidates[q] = candidates[q].intersection(H.neighbors(v))
                else:
                    newcandidates[q] = {w for w in candidates[q] if not H.has_edge(v,w)}
            if all(newcandidates.values()):
                yield from chooseRecursively(H, chosenVertices, availableVertices, chosenPoints, availablePoints, blockset, blockpatterns, vertexpatterns, newcandidates)
        # undo blockpatterns
        for b in blockset:
            blockpatterns[b].pop()
        # undo vertexpatterns
        for w in availableVertices:
            vertexpatterns[w].pop()
        chosenVertices.pop()
        availableVertices.add(v)
        chosenPoints.pop()
        availablePoints.add(p)
        H.delete_edge(v,p)
```

## Finding a cospectral mate

```python
def hasCospectralMate(G, method, certificate=False):
    Gcanonical = G.canonical_label()
    Q = switching_matrix[method]
    for Gamma in switching_subgraphs[method]:
        for C in allSwitchingSets(G, Gamma, blocks(Q)):
            if Gcanonical != switch(G, Q, C).canonical_label():
                if certificate:
                    return C, switch(G, Q, C)
                return True
    return False
```

## Finding all cospectral mates

```python
def allCospectralMates(G, method):
    representatives = [G.canonical_label()]
    Q = switching_matrix[method]
    for Gamma in switching_subgraphs[method]:
        for C in allSwitchingSets(G, Gamma, blocks(Q)):
            H = switch(G, Q, C).canonical_label()
            if not H in representatives:
                representatives.append(H)
                yield H
```
